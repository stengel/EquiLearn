```python
from learningAgents import ReinforceAlgorithm
from environmentModel import Model, AdversaryModes
from NeuralNetwork import NeuralNetwork
import torch
import torch.nn as nn
from torch.distributions import Categorical
import time
import matplotlib.pyplot as plt
```

```python
torch.manual_seed(0)
adversaryProbs=torch.zeros(len(AdversaryModes))
adversaryProbs[0]=1
game = Model(totalDemand = 400, 
               tupleCosts = (57, 71),
              totalStages = 25, adversaryProbs=adversaryProbs)
```

```python
st = time.time()
neuralNet=NeuralNetwork(lr = 0.00001, num_input= 2, num_actions=50, nn_dim = 200)
algorithm = ReinforceAlgorithm(game, neuralNet, numberBatches=1001, numberEpiPerBatch=1000)
algorithm.solver()
ft = time.time()
print("Time:", ft-st)
```

```python
import pandas as pd
learning = pd.DataFrame(algorithm.returns,columns=['entry'])
```

```python
learning.plot()
```
